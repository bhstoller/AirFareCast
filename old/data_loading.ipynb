```python
import pandas as pd
```

```python
df = pd.read_parquet("data/itineraries_snappy.parquet", engine="pyarrow", columns=["searchDate", "flightDate", "startingAirport", "destinationAirport","travelDuration", "isBasicEconomy", "isRefundable", "isNonStop", "totalFare", "seatsRemaining", "totalTravelDistance","segmentsDepartureTimeRaw", "segmentsArrivalTimeRaw", "segmentsAirlineCode", "segmentsCabinCode"]).iloc[:100]
```

```python
df.head()
```

```python
df.searchDate = pd.to_datetime(df.searchDate)
df.flightDate = pd.to_datetime(df.flightDate)
```

```python
extracted = df['travelDuration'].str.extract(r'PT(?:(\d+)H)?(?:(\d+)M)?').fillna(0).astype(int)

df['minutes'] = extracted[0].to_numpy() * 60 + extracted[1].to_numpy()
```

```python
df[df['totalTravelDistance'].isna()]
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')

df['totalTravelDistance'] = imputer.fit_transform(df[['totalTravelDistance']]).astype(int)
```

```python
df.head()
```

```python
df['segmentsDepartureTimeRaw'] = df['segmentsDepartureTimeRaw'].replace('', pd.NA)
df['segmentsArrivalTimeRaw'] = df['segmentsArrivalTimeRaw'].replace('', pd.NA)

df['segmentsDepartureTimeRaw'] = df['segmentsDepartureTimeRaw'].str.split('||').str[0]
df['segmentsArrivalTimeRaw'] = df['segmentsArrivalTimeRaw'].str.split('||').str[0]

df['segmentsDepartureTimeRaw'] = df['segmentsDepartureTimeRaw'].str.strip()
df['segmentsArrivalTimeRaw'] = df['segmentsArrivalTimeRaw'].str.strip()

df['segmentsDepartureTimeRaw'] = pd.to_datetime(df['segmentsDepartureTimeRaw'], errors='coerce', utc=True)
df['segmentsArrivalTimeRaw'] = pd.to_datetime(df['segmentsArrivalTimeRaw'], errors='coerce', utc=True)

df['departure_hour'] = df['segmentsDepartureTimeRaw'].dt.hour
df['arrival_hour'] = df['segmentsArrivalTimeRaw'].dt.hour

print(df[['segmentsDepartureTimeRaw', 'departure_hour', 'segmentsArrivalTimeRaw', 'arrival_hour']].head())
```

```python
average_seats_remaining = df[('seatsRemaining')].mean()
print(average_seats_remaining)
```

```python
mask_mixed_cabin = df['segmentsCabinCode'].str.contains(r'first.*\|\|.*(coach|business|premium coach)|'
                                                         r'coach.*\|\|.*(first|business|premium coach)|'
                                                         r'business.*\|\|.*(first|coach|premium coach)|'
                                                         r'premium coach.*\|\|.*(first|coach|business)',
                                                         regex=True)

percent_mixed_cabin = (mask_mixed_cabin.sum() / len(df)) * 100

print(f"Percentage of entries with mixed cabins: {percent_mixed_cabin:.2f}%")
```
